1. For each feature, write code to compute the average value, the min and max values, as well as its variance.
    
    Which features have the lowest and the highest average? Include the feature name and their average values for the features with the
    lowest and highest average.
    
    Which features have the lowest and the highest variance? Include the feature name and their variance values for the features with the
    lowest and highest variance.


```python
import pandas as pd

# load data
df = pd.read_csv("/Users/haderie/Downloads/housing/kc_house_data.csv")

# drop non-feature cols
features = df.drop(columns=['id', 'date', 'zipcode', 'price'])

# stats
stat = features.agg(['mean', 'min', 'max', 'var'])
print("Stat computation:\n", stat)

# lowest and highest average
lowest_avg_feature = stat.loc['mean'].idxmin()
lowest_avg_value = stat.loc['mean'].min()

highest_avg_feature = stat.loc['mean'].idxmax()
highest_avg_value = stat.loc['mean'].max()


print(f"Lowest average: {lowest_avg_feature} = {lowest_avg_value}")
print(f"Highest average: {highest_avg_feature} = {highest_avg_value}")

# lowest and highest variance
lowest_var_feature = stat.loc['var'].idxmin()
lowest_var_value = stat.loc['var'].min()

highest_var_feature = stat.loc['var'].idxmax()
highest_var_value = stat.loc['var'].max()

print(f"\nLowest variance: {lowest_var_feature} = {lowest_var_value}")
print(f"Highest variance: {highest_var_feature} = {highest_var_value}")
```

2. Compute the correlation coefficient of each feature with the response. Include a table with the correlation coefficient of each feature with the response. 
Which features are positively correlated (i.e., have positive correlation coefficient) with the response?

Which feature has the highest positive correlation with the response?


```python
df_features = df.drop(columns=['id', 'date', 'zipcode'])

corr = df_features.corr()

# correlation with response aka price
price_corr = corr.loc['price']

# table of correlation with features and price 
corr_table = price_corr.to_frame(name='corr_features_with_price')

print(corr_table.sort_values(by='corr_features_with_price', ascending=False))
```

All of the features are positively correlated with price, while some like 'yr_built', 'condition' and 'long' have small positive correlations compared with 'sqft_living' and 'grade' which have the highest positive correlation

sqft_living (Square footage of the apartments living space) has the highest positive correlation with price, which means that houses with larger living spaces tend to have higher prices as well.
